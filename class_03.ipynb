### if Statement


```python
# Basic if statement
age = 18

if age >= 18:
    print("You are eligible to vote")  # This will execute if condition is True
```

```python
# if statement with multiple lines
number = 10

if number > 5:
    print("Number is greater than 5")
    print("This is inside the if block")
    print("Multiple statements can be here")

print("This is outside the if block")
```

```python
# if statement with comparison
temperature = 25

if temperature > 30:
    print("It's hot outside!")

if temperature < 10:
    print("It's cold outside!")
```

### if-else Statement


```python
# if-else statement
age = 16

if age >= 18:
    print("You are eligible to vote")
else:
    print("You are not eligible to vote")  # This will execute if condition is False
```

```python
# if-else with number check
number = 7

if number % 2 == 0:
    print(f"{number} is even")
else:
    print(f"{number} is odd")
```

```python
# if-else with string comparison
username = "admin"

if username == "admin":
    print("Welcome, Administrator!")
else:
    print("Welcome, User!")
```

### if-elif-else Statement

```python
# if-elif-else statement
score = 85

if score >= 90:
    print("Grade: A")
elif score >= 80:
    print("Grade: B")  # This will execute
elif score >= 70:
    print("Grade: C")
elif score >= 60:
    print("Grade: D")
else:
    print("Grade: F")
```

```python
# if-elif-else with temperature
temperature = 25

if temperature > 30:
    print("It's very hot!")
elif temperature > 20:
    print("It's warm")  # This will execute
elif temperature > 10:
    print("It's cool")
else:
    print("It's cold!")
```

```python
# if-elif-else with age groups
age = 25

if age < 13:
    print("You are a child")
elif age < 20:
    print("You are a teenager")
elif age < 60:
    print("You are an adult")  # This will execute
else:
    print("You are a senior")
```

### Nested if Statements


```python
# Nested if statements
age = 20
has_license = True

if age >= 18:
    print("You are an adult")
    if has_license:
        print("You can drive")  # This will execute
    else:
        print("You need a license to drive")
else:
    print("You are a minor")
```

```python
# Nested if-elif-else
number = 15

if number > 0:
    print("Number is positive")
    if number % 2 == 0:
        print("And it's even")
    else:
        print("And it's odd")  # This will execute
elif number < 0:
    print("Number is negative")
else:
    print("Number is zero")
```

### Multiple Conditions with Logical Operators


```python
# Using 'and' operator
age = 25
has_license = True

if age >= 18 and has_license:
    print("You can drive")  # Both conditions must be True
else:
    print("You cannot drive")
```

```python
# Using 'or' operator
day = "Saturday"

if day == "Saturday" or day == "Sunday":
    print("It's weekend!")  # At least one condition must be True
else:
    print("It's a weekday")
```

```python
# Using 'not' operator
is_raining = False

if not is_raining:
    print("You can go outside")  # This will execute
else:
    print("Stay indoors")
```

```python
# Complex condition with multiple operators
age = 25
has_license = True
has_car = False

if age >= 18 and has_license:
    if has_car:
        print("You can drive your car")
    else:
        print("You can drive but you need a car")  # This will execute
else:
    print("You cannot drive")
```
