<!-- Data Types of Python -->
<!-- Str -->

```python
name = "Hello World"
print(name)
```

### Str Data Type

```python
full_name = "Hamza Rahim" # string "" , '' 
print(full_name) # printing the variable
print(type(full_name)) # printing the type of the variable
```

### Int Data Type 

```python
number = 10 # integer
print(number) # printing the variable
print(type(number)) # printing the type of the variable
```

### Float Data Type

```python
float_number = 10.5 # float
print(float_number) # printing the variable
print(type(float_number)) # printing the type of the variable
```

### Bool Data Type

```python
boolean_val = True # boolean
print(boolean_val) # printing the variable
print(type(boolean_val)) # printing the type of the variable
```

### Dict Data Type

```python
# dict
personal_info = {
    'name': "Hamza",
    'age': 30,
}
print(personal_info)
print(type(personal_info))
```

### List Data Type
- Ordered hoti hai
- Change (modify) ho sakti hai
- Duplicate values allowed

```python
my_list = [10, 20, 30, 40]

print(my_list)
print(type(my_list))   # List ka data type check
```

### Tuple

- Ordered hoti hai
- Change nahi ki ja sakti (immutable)
- Duplicate values allowed

```python
my_tuple = (10, 20, 30, 40)

print(my_tuple)
print(type(my_tuple))   # Tuple ka data type check
```

### Set 
- Unordered hota hai
- Duplicate values allowed nahi
- Items change nahi hotay, lekin set me items add/remove ho sakte hain

```python
my_set = {10, 20, 30, 40, 40}

print(my_set)           # 40 sirf ek dafa ayega
print(type(my_set))     # Set ka type
```
